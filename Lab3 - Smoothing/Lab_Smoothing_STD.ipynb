```python
'''
(1) Your own median filter algorithm
'''
def median_filter(img,win_width,win_height):
    newImg = np.copy(img)
    edgex = win_width // 2
    edgey = win_height // 2
    for x in range(edgex, newImg.shape[0] - edgex):
        for y in range(edgey, newImg.shape[1] - edgey):
            med = np.median(newImg[x:x+win_width, y:y+win_height])
            newImg[x+win_width//2,y+win_height//2]=med
    return newImg

```

```python
img = rgb2gray(io.imread('D:/CMP/third_Year/first_Semester/imageProcessing and computerVision/labs/lab2/imgs/bird.jpg'))

noisy_img = random_noise(img,mode="s&p",amount = 0.3)

```

```python

myMedImg = median_filter(noisy_img,7,7)
```

```python
show_images([noisy_img,myMedImg],['noisy Img','median filter'])
```

```python
'''
(2) skimage median filter
'''
filtered = median(noisy_img)
show_images([img,noisy_img,filtered],['Original','noisy Img','median filter'])
```

```python
'''
(3) Gaussion Filters
'''
noisy_img2 = random_noise(img,mode="localvar")
filt_img =  gaussian(noisy_img2,sigma=2)
show_images([img,noisy_img2,filt_img],['Original','noisy Img','gaussian filter'])
```
