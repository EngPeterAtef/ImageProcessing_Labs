```python
'''
(1) Your own median filter algorithm
'''
def median_filter(img,fiter_size):
    newImg = np.copy(img)
    edgex = fiter_size // 2
    edgey = fiter_size // 2
    for x in range(edgex, newImg.shape[0] - edgex):
        for y in range(edgey, newImg.shape[1] - edgey):
            med = np.median(newImg[x:x+fiter_size, y:y+fiter_size])
            newImg[x+fiter_size//2,y+fiter_size//2]=med
    return newImg

```

```python
img = rgb2gray(io.imread('../../lab2/imgs/bird.jpg'))
noisy_img = random_noise(img,mode="s&p",amount = 0.3)

```

```python
# 3x3 median filter (if the size increased the image will be more blury)
myMedImg = median_filter(noisy_img,3) 
```

```python
show_images([img,noisy_img,myMedImg],['Original','noisy Img','median filter'])
```

```python
'''
(2) skimage median filter
'''
filtered = median(noisy_img)
show_images([img,noisy_img,filtered],['Original','noisy Img','median filter'])
```

```python
'''
(3) Gaussion Filters
'''
noisy_img2 = random_noise(img,mode="localvar")
# increasing sigma means the neighbors of the pixel has more effect so the image will be more blury
filt_img =  gaussian(noisy_img2,sigma=2)
show_images([img,noisy_img2,filt_img],['Original','noisy Img','gaussian filter'])
```
